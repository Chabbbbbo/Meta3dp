```python
#### try문 위치 바꿔서 안읽히는 파일 패스하고 나머지 읽음

def extract_zip(zip_path,unzip_dir,rename=True):
    '''
    압축해제하는 함수
        -zip_path: 알집의 경로
        -unzip_dir: 알집해제하고자하는 폴더 경로
        -renmae: True 일 때, 알집파일명을 파일명에 앞에 붙임
    '''
    os.makedirs(unzip_dir,exist_ok=True)
    if '.zip' not in zip_path:
        print('zip파일이 아닙니다.')

    else: 
        with zipfile.ZipFile(zip_path) as zf:
            error_num = 0
            for file in zf.namelist():  # 루프를 돌며 하나의 파일을 가져옴
                try:
                    zip_name = zip_path.split('/')[-1].split('.zip')[0].split('_')[0].strip()
                    new_file_name = zip_name + '_' + file if rename else file
                    unzip_file_path = os.path.join(unzip_dir, new_file_name)
                    with open(unzip_file_path, "wb") as f:  # 저장되는 파일을 쓰기 모드로 오픈
                        # print('f: ',f)
                        f.write(zf.read(file))  # 압축된 파일을 읽어서 타깃 파일로 저장
                except Exception as e: 
                    # print(zip_name,'번 시편: Error',e)
                    error_num += 1
                    pass
            print(f'{zip_name}번 시편 =>> Error {error_num}개 제외 압축해제 완료')




# 원본 데이터셋 경로 설정
img_dir = '/content/drive/MyDrive/3DP공모전_On img'

# img_dir = raw_dataset_dir + '/image'
# label_dir = raw_dataset_dir + '/label'

# 압축해제 데이터셋 경로 설정
unzip_dataset_dir = '/content/dataset'
unzip_img_dir = unzip_dataset_dir + '/image'
# unzip_label_dir = unzip_dataset_dir + '/label'

# 전체 언집
# img_zip_path_list = [os.path.join(img_dir, img_zip) for img_zip in os.listdir(img_dir) if '.zip' in img_zip]
# 특정 넘버 언집시
unzip_num_list = ['19','20', '22', '23', '24', '26', '29']
img_zip_path_list = []
for img_zip in os.listdir(img_dir):
    if '.zip' in img_zip:
        if img_zip.split('_')[0] not in unzip_num_list:
            img_zip_path_list.append(os.path.join(img_dir, img_zip))

print(img_zip_path_list)
print(len(img_zip_path_list))
# img_zip_path_list = [img_zip for img_zip in os.listdir(img_dir) if ('.zip' and '17')  in img_zip]
# label_zip_path_list = [os.path.join(label_dir, label_zip) for label_zip in os.listdir(label_dir) if '.zip' in label_zip]

for img_zip_path in img_zip_path_list:
    extract_zip(img_zip_path,unzip_img_dir)

# 특정 파일 실행시 
# img_zip_path = '/content/drive/MyDrive/3DP공모전_On img/19_re_On_Axis_Images.zip'
# extract_zip(img_zip_path,unzip_img_dir)
```

### 1. 바운딩박스 도출해내는 함수 정의

```python
#### df 뽑아내는거 (사진저장 X)
def find_contours(img, threshold_percent):
    '''
    입력된 Image에서 Contous를 추출하여 max_area, w, h, ratio를 추출하는 함수
    각 Image에서 최대 Pixel을 기준으로 입력받은 Threshold percent만큼 임계값을 설정함

    예외처리)
    입력된 Image가 읽혀지지 않으면, Output모두 'Img_error'로 출력됨
    입력된 Image에서 threshold값으로 추출되는 Contous가 없는 경우, Output 모두 'None'으로 출력됨

    Input
     - img : 이미지의 경로
     - threshold_percent(float) : 임계값설정 % (0.nn 으로 입력)

     Output
     - max_area(float) : 찾은 Contous의 면적 (Contous가 여러개일 경우 최대값으로 설정) 
     - w(float, rount3) : Contous의 횡(가로) 값
     - h(float, round3) : Contous의 종(세로) 값
     - ratio(float, round3) : Contous의 종횡비 값
     img_bitwise_not_bgr2gray' is not defined
    '''
    
    try:
        img_bgr = cv2.imread(img)
        img_bitwise_not_bgr = cv2.bitwise_not(img_bgr)
        img_bitwise_not_bgr2gray = cv2.cvtColor(img_bitwise_not_bgr, cv2.COLOR_BGR2GRAY)
        max_pixel = img_bitwise_not_bgr2gray.max()
        threshold_value = max_pixel * threshold_percent 
#         ret, img_binary = cv2.threshold(img_bitwise_not_bgr, threshold_value,max_pixel,1)
        ret, img_binary = cv2.threshold(img_bitwise_not_bgr2gray, threshold_value,max_pixel,1)
        contours, hierarchy = cv2.findContours(img_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        cont_dict = dict()

        if len(contours) > 1: # contous 1개 이상
            for i in range(len(contours)):
                tmp_area = cv2.contourArea(contours[i])
                cont_dict[i] = tmp_area
            max_key = max(cont_dict, key=cont_dict.get)

        elif len(contours) == 1:
            max_key = 0
        else: # 잡힌 contous 없음
            max_key = 'contours_X'
        
        if max_key != 'contours_X':
            center, wh, rotate = cv2.minAreaRect(contours[max_key])
            max_area = float(cv2.contourArea(contours[max_key]))
            ratio =  float(wh[0]/wh[1])
            w = round(wh[0], 3)
            h = round(wh[1], 3)
            ratio = round(ratio, 3)
            contours = contours[max_key]
        else:
            max_area = w = h = ratio = 'None'
    except Exception as e:
        # print(img.split('/')[-1], ' error msg: ',e)
#         max_area , w , h , ratio = None
        max_area = w = h = ratio = 'Img_error'

    return max_area, w, h, ratio
```

```python
'''
= 속도개선 =
기존 img 47994개(28번시료) 완료 시간 : 516.2216429710388
개선 img 47994개(28번시료) 완료 시간 : 75.02418303489685 
[0]리스트 img 47994개 완료 시간 :  46.55679893493652 (1s 1,032.1개)
===========
시편 312967개(시편 7개) 완료 시간 :  1213.9864437580109 (1s 257.8개 )
시편 268841개(시편 7개) 완료 시간 :  655.6749081611633
-> 출력되는 Error 값이 많아서 그런거 같음.
'''

import os
import pandas as pd
import warnings
import time
import numpy as np

warnings.filterwarnings('ignore')

start = time.time()
root_dir = '/content/dataset/image'

# 특정 샘플 넘버만 진행시
# sample_nums = ['29', '30', '31']
# samples_1 = [f for f in os.listdir(root_dir) if f.split('_')[0] in sample_nums]

# 압축푼 전체 넘버 진행
sample_nums = set([file.split('_')[0] for file in os.listdir(unzip_img_dir) ])
samples = [f for f in os.listdir(root_dir) if f.split('_')[0] in sample_nums]

# 여기서 먼저 빈 list를 만들어주고 시작함
samples = os.listdir('/content/dataset/image')
array_list = [[0] for i in range(len(samples))]

# 조건 설정
threshold_percent = 0.1
columns= ['specimen_no', 'Layer No.','contours_threshold', 'area','length', 'width', 'aspect_ratio']

num = 0
for idx, temp_file in enumerate(samples):
    # print('   = > 진행중인 파일', temp_file)
    sample_num = temp_file.split('_')[0]
    temp_file_path = root_dir+'/'+temp_file
    max_area, w, h, ratio = find_contours(temp_file_path,threshold_percent)
    data = [sample_num,temp_file,threshold_percent,str(max_area), str(w),str(h), str(ratio)]
    # array_list = np.append(array_list,data,axis=0)
    array_list[num] = data 
    num +=1
    # temp_df = temp_df.append(pd.Series(data, index=temp_df.columns), ignore_index=True)
    # temp_df.to_csv('/content/'+sample_num+'_on_bbox_info.csv')
    if num % 50000 == 0:
        print(f'{num}개 완료')

df = pd.DataFrame(array_list, columns = columns)
df.to_csv(f'/content/img_info{sample_nums}_thres_10.csv',mode ='w', header=True, encoding='utf-8-sig' )
end = time.time()
print(f'시편 {num}개 완료 시간 : ', end - start) # 시편 1개 759.3309307098389, img 184233개 완료 시간 :  8446.08764743805, img 102424개 완료 시간 :  2112.132268190384
```
