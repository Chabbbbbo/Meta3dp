```python
import pandas as pd
import os
import time
```

```python
###### 수정 후 

root_path = "C:\\Users\\Chaaaaabbo\\Desktop\\3DP\\230201_신규데이터진행"
time_dict_ = {}
time_dict = {}
for i in range(1,49):
    start = time.time()
    file_path = os.path.join(root_path, f'0_threshold_10\\img_info{i}_thres_10.csv')
    data = pd.read_csv(file_path)
    change_num = 0
    none_num = 0
    # data = pd.read_csv(file_path, index_col=0) 'Unnamed: 0' 안넣고 싶을때
    for j in range(len(data)):
        try:
            length = float(data[data['Unnamed: 0'] == j]['length'])
            width = float(data[data['Unnamed: 0'] == j]['width'])
            if width > length:
                long = width
                short = length
                data.loc[data['Unnamed: 0'] == j, "length"] = long
                data.loc[data['Unnamed: 0'] == j, "width"] = short
                data.loc[data['Unnamed: 0'] == j, "aspect_ratio"] = round(long/short,3)
                change_num += 1
            else: # length가 width와 같거나 크면 넘어감
                pass

        except:
            none_num += 1
            pass

    data.rename(columns = {'length':'long','width':'short'},inplace=True)
    time_ = time.time() - start
    time_dict_["총 갯수"] = len(data)
    time_dict_["변경 갯수"] = change_num
    time_dict_["try 갯수"] = len(data) - none_num
    time_dict_["None 갯수"] = none_num
    time_dict_["시간"] = time_
    time_dict[i] = time_dict_
    output_path = os.path.join(root_path, f'0_threshold_10_ls\\img_info_0_{i}_thres_10_ls.csv')
    data.to_csv(output_path, index = True)
print(f'{i} 완료, {len(data)}개 중 {len(data) - none_num}개 데이터 {change_num}개 변경 소요시간 {time_}')
```
